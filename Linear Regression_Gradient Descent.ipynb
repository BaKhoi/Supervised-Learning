```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8,6) #width and height of figure

# Preprocessing Input data
data = pd.read_csv('data_LR.csv')
X = data.iloc[:, 0]#all rows, 1st column
Y = data.iloc[:, 1]#all rows, 2nd column
plt.scatter(X, Y)
plt.show()
```

```python
# Building the model
w = 0
b = 0

L = 0.0001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent, this can be 100, 200, ...

n = len(X) # Number of elements in X

#Gradient Descent
for i in range(epochs):
    Y_pred = w*X + b
    D_w = (1/n)*sum((Y_pred - Y)*X)#Subtract 2 lists
    D_b = (1/n)*sum(Y_pred - Y)
    #update
    w = w - L*D_w
    b = b - L*D_b


print(w,b) 

#1.476839539354219 0.24477587515819155
```

```python
# Making predictions
Y_pred = w*X + b
df_preds = pd.DataFrame({'Actual': Y.squeeze(), 'Predicted': Y_pred.squeeze()})
print(df_preds)

plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```
